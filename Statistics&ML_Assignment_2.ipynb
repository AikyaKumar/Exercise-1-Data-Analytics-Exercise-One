```python
# Import necessary libraries
import numpy as np  # For numerical computations
import pandas as pd  # For data manipulation and analysis
import matplotlib.pyplot as plt  # For creating visualizations

from sklearn.datasets import load_iris  # To load the Iris dataset

# Load the Iris dataset as a pandas DataFrame
data = load_iris(as_frame=True)
df = data.frame  # Assign the DataFrame to 'df'
print(df.head())  # Display the first few rows of the DataFrame

# Take a random sample of 30 rows from the DataFrame
random_sample = df.sample(n=30)
print("\n Random Samples: \n", random_sample.head())  # Display the first few rows of the random sample

# Define the sample size for calculating sample means
sample_size = [100]

# Calculate sample means of sepal length for 100 samples of size 100
sample_means_sepal_length = {size: [df['sepal length (cm)'].sample(size).mean() for _ in range(100)] for size in sample_size}
print("\n Sample Means: \n", sample_means_sepal_length)  # Print the calculated sample means

# Create a histogram to visualize the distribution of sample means
plt.figure(figsize=(15, 5))  # Set the figure size
for size, means in sample_means_sepal_length.items():
    plt.hist(means, bins=10, alpha=0.5, label=f'Sample Size {size}')  # Create histogram for each sample size

plt.legend()  # Display the legend
plt.title('\n Distribution of Sample Means of Sepal Length')  # Set the title of the plot
plt.xlabel('Sample Mean')  # Set the x-axis label
plt.ylabel('Frequency')  # Set the y-axis label
plt.show()  # Display the plot
```

```python

# Calculate the sampling interval (step) for systematic sampling
# We want a sample size of approximately 20% of the original data
step = int(len(df) // (len(data.frame) * 0.2))

# Create a systematic sample by selecting rows at regular intervals (step)
systematic_sample = df.iloc[::step]

# Print the first few rows of the systematic sample to inspect it
print("\n Systematic Sample: \n", systematic_sample.head())

# Plotting the distributions
plt.figure(figsize=(15, 5))

# Plotting the Original Dataset
plt.subplot(1, 2, 1)  # Create a subplot (1 row, 2 columns, select the 1st subplot)
plt.hist(df, bins=10, edgecolor='black')  # Create a histogram of the original data
plt.title('Original Dataset')  # Set the title of the plot

# Plotting the Sample Dataset
plt.subplot(1, 2, 2)  # Select the 2nd subplot in the 1x2 grid
plt.hist(systematic_sample, bins=10, edgecolor='black')  # Create a histogram of the sample data
plt.title('Systematic Sample')  # Set the title of the plot

plt.tight_layout()  # Adjust the spacing between subplots
plt.show()  # Display both plots
```
